```python
import os
import numpy as np
import pandas as pd

import utils
```

```python
test_index = pd.read_csv('data/test_public.csv.gz', sep=',', usecols=[0])
test_index = test_index['id']
```

```python
df_preds_mlp = pd.read_csv('preds/hepmlp_test_preds.csv')
df_preds_cat = pd.read_csv('preds/catboost_test_preds.csv')
df_preds = pd.concat([df_preds_mlp, df_preds_cat], axis=1)
df_preds.head()
```

```python
y_pred = df_preds.values
y_pred_wgth = 0.4 * y_pred[:, :5].mean(axis=1) + 0.6 * y_pred[:, 5:].mean(axis=1)
```

```python
pd.DataFrame(data={"prediction": y_pred_wgth}, index=test_index).to_csv(
        "submission_public_weighted.csv", index_label=utils.ID_COLUMN)
```
